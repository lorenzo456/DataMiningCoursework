```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```python
x_train_gr_smpl = pd.read_csv("x_train_gr_smpl.csv",delimiter=",",dtype=np.uint8)
```

```python
row_1 = x_train_gr_smpl.iloc[0]
row_1413 = x_train_gr_smpl.iloc[1413] 
row_3290 = x_train_gr_smpl.iloc[3290]
row_3695 = x_train_gr_smpl.iloc[3695]
row_5015 = x_train_gr_smpl.iloc[5015]
row_7115 = x_train_gr_smpl.iloc[7115]
row_9280 = x_train_gr_smpl.iloc[9280]
row_10060 = x_train_gr_smpl.iloc[10060]
row_10300 = x_train_gr_smpl.iloc[10300]
row_12400 = x_train_gr_smpl.iloc[12400]
```

```python
def convert2matrix(row):
    row = np.reshape([row],(48,48),order="c")
    return row   
```

```python
row_1 = convert2matrix(row_1)
row_1413 = convert2matrix(row_1413)
row_3290 = convert2matrix(row_3290)
row_3695 = convert2matrix(row_3695)
row_5015 = convert2matrix(row_5015)
row_7115 = convert2matrix(row_7115)
row_9280 = convert2matrix(row_9280)
row_10060 = convert2matrix(row_10060)
row_10300 = convert2matrix(row_10300)
row_12400 = convert2matrix(row_12400)
```

```python
fig, axs = plt.subplots(nrows=2,ncols=5,figsize=(13, 5),subplot_kw={'xticks': [], 'yticks': []})
axs[0,0].imshow(row_1)
axs[0,0].set_title("0")
axs[0,1].imshow(row_1413)
axs[0,1].set_title("1")
axs[0,2].imshow(row_3290)
axs[0,2].set_title("2")
axs[0,3].imshow(row_3695)
axs[0,3].set_title("3")
axs[0,4].imshow(row_5015)
axs[0,4].set_title("4")
axs[1,0].imshow(row_7115)
axs[1,0].set_title("5")
axs[1,1].imshow(row_9280)
axs[1,1].set_title("6")
axs[1,2].imshow(row_10060)
axs[1,2].set_title("7")
axs[1,3].imshow(row_10300)
axs[1,3].set_title("8")
axs[1,4].imshow(row_12400)
axs[1,4].set_title("9")
```

```python
threshold_value = 124
def binarize (row):
    row = (row<threshold_value) * 255
    return row
```

```python
bin_row_1 = binarize(row_1)
bin_row_1413 = binarize(row_1413)
bin_row_3290 = binarize(row_3290)
bin_row_3695 = binarize(row_3695)
bin_row_5015 = binarize(row_5015)
bin_row_7115 = binarize(row_7115)
bin_row_9280 = binarize(row_9280)
bin_row_10060 = binarize(row_10060)
bin_row_10300 = binarize(row_10300)
bin_row_12400 = binarize(row_12400)
```

```python
fig, axs = plt.subplots(nrows=2,ncols=5,figsize=(13, 5),subplot_kw={'xticks': [], 'yticks': []})
axs[0,0].imshow(bin_row_1,cmap="gray")
axs[0,0].set_title("0")
axs[0,1].imshow(bin_row_1413,cmap="gray")
axs[0,1].set_title("1")
axs[0,2].imshow(bin_row_3290,cmap="gray")
axs[0,2].set_title("2")
axs[0,3].imshow(bin_row_3695,cmap="gray")
axs[0,3].set_title("3")
axs[0,4].imshow(bin_row_5015,cmap="gray")
axs[0,4].set_title("4")
axs[1,0].imshow(bin_row_7115,cmap="gray")
axs[1,0].set_title("5")
axs[1,1].imshow(bin_row_9280,cmap="gray")
axs[1,1].set_title("6")
axs[1,2].imshow(bin_row_10060,cmap="gray")
axs[1,2].set_title("7")
axs[1,3].imshow(bin_row_10300,cmap="gray")
axs[1,3].set_title("8")
axs[1,4].imshow(bin_row_12400,cmap="gray")
axs[1,4].set_title("9")
```

```python
def adaptiveThresh(row):
    row = cv2.adaptiveThreshold(row,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)
    return row
```

```python
adbin_row_1 = adaptiveThresh(row_1)
adbin_row_1413 = adaptiveThresh(row_1413)
adbin_row_3290 = adaptiveThresh(row_3290)
adbin_row_3695 = adaptiveThresh(row_3695)
adbin_row_5015 = adaptiveThresh(row_5015)
adbin_row_7115 = adaptiveThresh(row_7115)
adbin_row_9280 = adaptiveThresh(row_9280)
adbin_row_10060 = adaptiveThresh(row_10060)
adbin_row_10300 = adaptiveThresh(row_10300)
adbin_row_12400 = adaptiveThresh(row_12400)
```

```python
fig, axs = plt.subplots(nrows=2,ncols=5,figsize=(13, 5),subplot_kw={'xticks': [], 'yticks': []})
axs[0,0].imshow(adbin_row_1,cmap="gray")
axs[0,0].set_title("0")
axs[0,1].imshow(adbin_row_1413,cmap="gray")
axs[0,1].set_title("1")
axs[0,2].imshow(adbin_row_3290,cmap="gray")
axs[0,2].set_title("2")
axs[0,3].imshow(adbin_row_3695,cmap="gray")
axs[0,3].set_title("3")
axs[0,4].imshow(adbin_row_5015,cmap="gray")
axs[0,4].set_title("4")
axs[1,0].imshow(adbin_row_7115,cmap="gray")
axs[1,0].set_title("5")
axs[1,1].imshow(adbin_row_9280,cmap="gray")
axs[1,1].set_title("6")
axs[1,2].imshow(adbin_row_10060,cmap="gray")
axs[1,2].set_title("7")
axs[1,3].imshow(adbin_row_10300,cmap="gray")
axs[1,3].set_title("8")
axs[1,4].imshow(adbin_row_12400,cmap="gray")
axs[1,4].set_title("9")
```
